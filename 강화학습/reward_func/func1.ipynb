```python
import math

def reward_function(params):
    
    ############## [reward1] all_wheels_on_track #############
    all_wheels_on_track = params['all_wheels_on_track']
    
    if all_wheels_on_track :
        reward1 = 0.1 ##set value
    else:
        reward1 = 0 ##set value

    ############## [reward2] speed #############
    speed = params['speed']
    
    if speed > 2 :
        reward2 = 0.3 ##set value
    elif speed > 1.5 :
        reward2 = 0.2 ##set value
    elif speed > 1 :
        reward2 = 0.1 ##set value
    else:
        reward2 = 0  ##set value

        
    ############## [reward3] distance_from_center #############
    distance_from_center = params['distance_from_center']
    track_width = params['track_width']

    if distance_from_center < 0.1 * track_width :
        reward3 = 0.2 ##set value
    elif distance_from_center < 0.5 * track_width :
        reward3 = 0.1 ##set value 
    else:
        reward3 = 0 ##set value
        
        
    ############## [reward4] steering_angle #############
    abs_steering = abs(params['steering_angle'])
    
    if abs_steering < 5 :
        reward4 = 0.4 ##set value
    elif abs_steering < 10 :
        reward4 = 0.3 ##set value
    elif abs_steering < 15 :
        reward4 = 0.2 ##set value
    elif abs_steering < 20 :
        reward4 = 0.1 ##set value
    else :
        reward4 = 0 ##set value
    
    
    ############## [reward5] direction_diff #############
    waypoints = params['waypoints']
    closest_waypoints = params['closest_waypoints']
    heading = params['heading']
    
    # 현재 에이전트와 가장 가까운 waypoint의 위치를 가져옵니다.
    next_point = waypoints[closest_waypoints[1]]
    prev_point = waypoints[closest_waypoints[0]]

    # track_direction 을 계산 
    track_direction = math.atan2(next_point[1] - prev_point[1], next_point[0] - prev_point[0])
    track_direction = math.degrees(track_direction)  

    # 각의 차이 (direction_diff) 계산
    direction_diff = abs(track_direction - heading)
    if direction_diff > 180:
        direction_diff = 360 - direction_diff
    
    # 보상 결정
    if direction_diff < 5 :
        reward5 = 0.4 ##set value
    elif direction_diff < 10 :
        reward5 = 0.3 ##set value
    elif direction_diff < 15 :
        reward5 = 0.2 ##set value
    elif direction_diff < 20 :
        reward5 = 0.1 ##set value
    else :
        reward5 = 0 ##set value 
        
        
    ########### [final reward] ############ 
    reward = reward1 * reward2 * reward3 * reward4 * reward5
    
    return float(reward)
```

```python
## 중앙간의 거리
import math

def reward_function(params):
  '''
  Example of rewarding the agent to follow center line
  '''
  
  # Read input parameters
  track_width = params['track_width'] ## 트랙 넓이 
  distance_from_center = params['distance_from_center'] ## 트랙의 중앙과 agent 사이의 거리
  all_wheels_on_track = params['all_wheels_on_track'] ## 바퀴가 모두 트랙에 있을 경우(true), 1개라도 트랙에 벗어나면 false
  abs_steering = abs(params['steering_angle']) ## 회전각
  is_left_of_center = params['is_left_of_center'] ## 차량이 트랙 중앙에서 왼쪽에 있을 경우(true) 아닐 경우 false
  speed = params['speed'] ## 차량의 스피드
  waypoint = params['waypoints'] ## 각 구간별 waypoint
  close_point = params['closest_waypoints'] ## 차량의 앞바퀴와 가까운 waypoint
  heading = params['heading'] ## 차량의 진행방향

  
  next_point = waypoint[close_point[1]] ## 다음 waypoint
  prev_point = waypoint[close_point[0]] ## 이전 waypoint  
  
  # Calculate 3 markers that are at varying distances away from the center line
  marker_1 = 0.2 * track_width ## 트랙의 20%
  marker_2 = 0.35 * track_width ## 트랙의 35%
  marker_3 = 0.6 * track_width ## 트랙의 60%
  
  reward = 1e-3

  if all_wheels_on_track and (0.5 * track_width - distance_from_center) >= 0.05: ## 만약 4바퀴가 모두 트랙에 있으며, 트랙의 50%가 트랙중앙과의 거리보다 0.05 이상일 때 reward를 10로 받는다
    reward = 10
  
  # Give higher reward if the car is closer to center line and vice versa
  if distance_from_center <= marker_1: ## 차량과 agent사이의 거리가 20%이하면 reward 1
    reward = 10 
  elif distance_from_center <= marker_2: ## 차량과 agent사이의 거리가 20%와 35% 사이면 reward 0.5
    reward = 7
  elif distance_from_center <= marker_3: ## 차량과 agent사이의 거리가 35%와 60% 사이면 reward 0.1
    reward = 5
  else:
    reward = 1e-3 ## 차량과 agent사이의 거리가 50%이상이면 거의 0




  if (is_left_of_center == True):
    if (speed > 2.5): ## 스피드를 받아오는 함수를 사용해서 스피드를 2 이상이 될 경우 reward = 0.3, 아닐경우는 1
      reward = 5
    else:
      reward = 10

  track_direction = math.atan2(next_point[1] - prev_point[1], next_point[0] - prev_point[0]) ## next_point 와 prev_point 거리 구하기
  track_direction = math.degrees(track_direction - heading) ## 각도로 표현하기 위해 degree로 각도 변환

  direction_diff = abs(track_direction - heading) 
  if(direction_diff > 180): ## 180도가 넘어가면 음수로 되기 때문에 절대값하기
    direction_diff = 360 - direction_diff
  
  if(direction_diff <=8): ## 중앙선 방향과 트랙의 방향이 8도보다 적을 경우 reward = 1
    reward = 10
  else:
    reward = 3
  
  # Steering penality threshold, change the number based on your action space setting
  ABS_STEERING_THRESHOLD = 20 ## 회전각도의 임계값
  
  # Penalize reward if the car is steering too much
  if abs_steering > ABS_STEERING_THRESHOLD: ## 회전스티어링이 회전각도 20도가 넘어갈 경우 reward를 30%를 줄인다.
    reward *= 0.7

  return float(reward)
```
