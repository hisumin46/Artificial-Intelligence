#### ndarray와 DataFrame
- DataFrame과 리스트, 딕셔너리, 넘파이 ndarray **상호변환**
- DataFrame은 리스트, 딕셔너리, ndarray 등 변환가능
- 인공지능 라이브러리 ScikitLearn은 대부분 ndarray를 인자로 받음
- 상호간의 변환을 할 일이 많음

##### ndarray, list → DataFrame 
- 컬럼을 가지고 있어 데이터 핸드링이 비교적 쉬움
- 번호나 컬럼명 지정(지정하지 않으면 자동으로 컬럼명 할당)
- df의 생성인자 data는 값을 입력받고 생성인자 columns는 컬럼명 리스트 입력받아 쉽게 DataFrame을 생성
- 2차원 이하 데이터들만 df로 변환가능

````python
import numpy as np

# col이 1개의 경우
col_name1=['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
````

````python
# 2차원 형태의 데이터 변환

# 3개의 칼럼명이 필요
col_name2=['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1, 2, 3], [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape )

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
````

##### 딕셔너리 → DataFrame
- Key는 컬럼명 Value는 키에 해당하는 컬럼데이터로 변환
- Key의 경우 문자열, 값의경우 리스트 or ndarray 형태로 딕셔너리 구성

````python
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
````

##### 딕셔너리 → ndarray
- 머신러닝 lib가 기본 데이터형으로 ndarray를 사용하기 때문에 다시 변환해야 할 때가 빈번하게 발생
- 데이터 프레임 객체의 valuse를 이용하여 쉽게 할 수 있음

````python
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
````

##### DataFrame → list, Dictionary
- 리스트로 변환은 values로 얻은 ndarray에 tolist()호출
- 딕셔너리로 변환은 df의 to_dict() 메서드를 호출(인자로 list 입력시 다시 리스트형으로 변환)

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3) 
````

#### 데이터 셀렉션과 필터링
- 판다스의 데이터 셀렉션과 필터링은 넘파이와 비슷하여 사용할 때 헷갈리기 쉬워 주의
- 넘파이 : [ ] 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱 가능
- 판다스 : ix[], lioc[], loc[]연산자 사용
- 데이터 프레임 바로 뒤에 오는 [ ]는 컬럼만 지정할 수있는 컬럼 지정 연산자로 이해

##### [ ] 연산자
- 데이터프레임 바로 뒤에오는 [ ]는 컬럼만 지정할수있는 **컬럼 지정 연산자**
- `DatatFrame['컬럼명']` : '컬럼명'에 해당하는 컬럼 데이터 추출
- `DatatFrame[['컬럼명', .... , '컬럼명']]` : 컬럼명 리스트에 해당하는 모든 커럶의 데이터 추출
- `DatatFrame[0, 0]` : 오류
- [ ] 연산자는 넘파이나 시리즈의 [ ]와 다름
- DataFrame의 [ ]내 입력 값은 컬럼명(또는 컬럼 리스트)을 지정해서 컬럼 지정 연산에 사용하거나 불린 인덱스 용도
- 슬라이싱 연산으로 데이터 추출은 안하는 것이 좋음

##### DataFrame - iloc[ ] 연산자
- iloc[]는 위치 기반 인덱싱으로 행과 열 값으로 integer또는 integer 형의 슬라이싱, 팬시 리스트 값을 입력해야됨
- 위치기반 인덱싱만 가능하며 행, 열 위치 값으로 정수형 값을 지정하여 데이터 출력

````python
data = {
  'Name' : ['Chumin', 'Eunkung', 'Jinwoong', 'Soobeom'], 
  'Year' : [2011, 2016, 2015, 2015],
  'Gender' : ['M', 'F', 'M', 'M']
}

data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])
data_df
````

````python
# 첫번대 행, 열 추출
data_df.iloc[0, 0]
````

##### DataFrame - loc[ ] 연산자
- loc[]는 명칭 기반 인덱싱으로 행 위치에는 index값, 열 위치에는 컬럼명 입력
- 슬라이싱 기호 **":"** 적을때 유의점
 - 일반적으로 슬라이싱을 `시작:종류` 로 지정하면 `시작 값 ~ 종료 -1`까지의 범위를 의미
 - 하지만 loc[]에 슬라이싱 기호를 적용하면 종료값까지 출력
 - 이유는 명칭은 숫자 형이 아닐 수 있으므로 -1을 할 수 없음
- 즉 명칭 기반 인덱싱만 가능하며 행 위치에 인덱스가 오고 열 위치에 컬럼명을 지정

````python
data_df.loc['one', 'Name']
````

````python
print('### 위치 기반 lioc 슬라이싱 ### \n', data_df.iloc[0:1, 0], '\n')
print('### 명칭 기반 lioc 슬라이싱 ### \n', data_df.loc['one' : 'two', 'Name'])
````
