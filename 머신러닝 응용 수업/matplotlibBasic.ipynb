```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,10], 'o')
plt.axis()
plt.show()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,10])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,10], label='Price($)')
plt.plot([1,2,3,4], [1,3,8,10], label='Price0 ($)')
plt.xlabel('x')
plt.ylabel('y')
# loc는 위치
plt.legend(loc=(0.5, 1.0))
# 숫자가 아니어도 가능
plt.legend(loc='lower right')
plt.show()
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```
