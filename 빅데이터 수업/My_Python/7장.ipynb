##### 회귀분석 - p 값으로 판단
- p값을 기준으로 긍정과 부정을 분석한다
- 만약 기준이 0.05라면
- 0.04가 나와서 몰입이 중독에 영향을 미친다!라고 할수있음
- 0.06이 나오면몰입이 중독에 영향을 미치지 않는다!라고 할수있음
- 0.05이 나오면몰입이 중독에 영향을 미친다!라고 할수있음
- 0.054이 나오면몰입이 중독에 영향을 미친다!라고 할수있음
- 0.056이 나오면몰입이 중독에 영향을 미치지 않는다!라고 할수있음 - 애매할땐 걍 0.56->0.05로 조진다 -> 채택으로 조진다


```python
from scipy import stats
from statsmodels.formula.api import ols, glm
red_wind_q = wine.loc[wine['type'] == 'red', 'quality']
white_wind_q = wine.loc[wine['type'] == 'white', 'quality']
stats.ttest_ind(red_wind_q, white_wind_q, equal_var=False)
```

```python
wine.columns
```

```python
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
regressioin_res = ols(Rformula, data=wine).fit()
regressioin_res.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
sns.distplot(red_wind_q, kde=True, label='red wind')
sns.distplot(white_wind_q, kde=True, label='white wine')
plt.title('Quality of Wind Type')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
others = list(set(wine.columns).difference(set(["quality", "fixed_acidity"])))
p, resids = sm.graphics.plot_partregress("quality", "fixed_acidity", others, data=wine, ret_coords=True)
plt.show()
```

```python
fig = plt.figure(figsize=(8,13))
sm.graphics.plot_partregress_grid(regressioin_res, fig=fig)
plt.show()
```

#### 타이타닉 분류 예측

```python
import seaborn as sns
import pandas as pd
titanic = sns.load_dataset("titanic")
titanic
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(1,2, figsize=(10, 5))
titanic['survived'][titanic['sex'] == 'male'].value_counts().plot.pie(explode = [0,0.1], autopct = '%1.1f%%', ax = ax[0], shadow = True)
titanic['survived'][titanic['sex'] == 'female'].value_counts().plot.pie(explode = [0,0.1], autopct = '%1.1f%%', ax = ax[1], shadow = True)
ax[0].set_title('Survived - Famale')
ax[1].set_title('Survived - Male')
plt.show()
```

```python
sns.countplot('pclass', hue='survived', data=titanic)
plt.title('Pclass vs Survived')
plt.show()
```

```python
# 상관관계
titanic_corr = titanic.corr(method='pearson')
titanic_corr
```
